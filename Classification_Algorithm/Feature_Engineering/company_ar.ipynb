```python
import pandas as pd
import numpy as np
```

```python
path = "../Data/Online_Data/trainning_data/corporate_attributes/"
outpath = "../Processed_Data/"
```

```python
company_ar= pd.read_csv(path+"company_ar.csv")
```

```python
company_ar
```

```python
company_ar = company_ar.drop(["BUSST","ANCHEDATE"],axis=1)
company_ar = company_ar.fillna(0)
```

```python
company_ar
```

```python
numerical_columns = ['EMPNUM','WOMEMPNUM']
```

```python
def get_numerical_df(df, columns):
    keep_columns = columns + ['entid']
    df_numerical = df[keep_columns].groupby("entid")
    df_numerical = df_numerical.mean()
    df_numerical = df_numerical.reset_index()
    return df_numerical
```

```python
df_numerical = get_numerical_df(company_ar, numerical_columns)
```

```python
df_numerical["women_ratio"]  = 0
df_numerical.loc[df_numerical["WOMEMPNUM"] != 0, "women_ratio"] = df_numerical["WOMEMPNUM"]/df_numerical["EMPNUM"]
df_numerical.loc[df_numerical["women_ratio"] >= 1, "women_ratio"] = 1
df_numerical.drop(["EMPNUM","WOMEMPNUM"],axis=1)
```

```python
df_numerical.to_csv(outpath+"company_ar_processed.csv",index=False)
```
