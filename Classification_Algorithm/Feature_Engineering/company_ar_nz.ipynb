```python
import pandas as pd
import numpy as np
```

```python
path = "../Data/Online_Data/trainning_data/corporate_attributes/"
outpath = "../Processed_Data/"
```

```python
company_ar_nz = pd.read_csv(path+"company_ar_nz.csv")
```

```python
company_ar_nz = company_ar_nz.fillna({'FARNUM':0, 'ANNNEWMEM': 0, 'ANNREDMEM': 0})

company_ar_nz = company_ar_nz.groupby('entid').mean()
company_ar_nz = company_ar_nz.drop(["ANCHEYEAR"],axis=1)
```

```python
company_ar_nz
```

```python
company_ar_nz["concentration(ESG)"] = 0
company_ar_nz.loc[company_ar_nz["MEMNNUM"] != 0, "concentration(ESG)"] = company_ar_nz["FARNUM"]/company_ar_nz["MEMNNUM"]  
company_ar_nz.loc[company_ar_nz["concentration(ESG)"] >= 1, "concentration(ESG)"] = 1

company_ar_nz["Turnover of farmer"] = 0
company_ar_nz["Turnover of farmer"] = company_ar_nz["ANNNEWMEM"] -company_ar_nz["ANNREDMEM"]
```

```python
company_ar_nz.to_csv(outpath+"company_ar_nz_processed.csv")
```
