```python
import pandas as pd
import numpy as np
```

```python
path = "../Data/Online_Data/trainning_data/corporate_attributes/"
outpath = "../Processed_Data/"
```

```python
tax_year = pd.read_csv(path+"tax_year.csv")
```

```python
tax_year_num = tax_year.drop(["islatest"],axis=1).copy()
tax_year_num = tax_year_num.groupby(['entid',"year"]).mean()
tax_year_num = tax_year_num.groupby(["entid"]).mean()
tax_year_num
```

```python
tax_year_num["profit_margin"] = tax_year_num["total_profit"]/tax_year_num["revenue"]
tax_year_num["operating_profit_margin"] = tax_year_num["operation_profit"]/tax_year_num["revenue"]
tax_year_num = tax_year_num.drop(["assets"],axis=1)

list_sum = ["sales_expense","G_expense","asset_devaluation_losses","payrol_expense","welfare_expenses","education_expenses",\
           "union_funds","ad_expenses","entertainment_expenses","assets_loss"]
fildict = {}
for i in range(len(list_sum)):
    fildict[list_sum[i]] = 0

tax_year_num = tax_year_num.fillna(0)
tax_year_num["other_expense"] = 0

for i in range(len(list_sum)):
    tax_year_num["other_expense"] += tax_year_num[list_sum[i]]
```

```python
tax_year_num = tax_year_num.drop(["sales_expense","G_expense","asset_devaluation_losses","payrol_expense","welfare_expenses","education_expenses",\
           "union_funds","ad_expenses","entertainment_expenses","assets_loss"],axis=1) 
```

```python
tax_year_num["net_asset"] = tax_year_num["assets_re"] - tax_year_num["asset_devaluation_allowance"]
tax_year_num["main_biz_profit_margin"] = tax_year_num["main_business_income"] - tax_year_num["main_business_cost"]/tax_year_num["main_business_income"]
```

```python
tax_year_num = tax_year_num.drop(["asset_devaluation_allowance","income_tax_adjuste","income_taxable","incometax_taxable","incometax_amount","incometax_amount"],axis=1) 
```
