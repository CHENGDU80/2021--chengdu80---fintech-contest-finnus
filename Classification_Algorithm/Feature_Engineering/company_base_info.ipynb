```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
path = "../Data/Online_Data/trainning_data/corporate_attributes/"
outpath = "../Processed_Data/"
```

```python
company_base_info = pd.read_csv(path+"company_base_info.csv")
```

```python
company_base_info 
```

```python
company_base_info["honor_num"]  = company_base_info["honor_num"].fillna(0)
company_base_info["illegad_num"]  = company_base_info["illegad_num"].fillna(0)
company_base_info["yea_sub"]  = company_base_info["yea_sub"].fillna(0)
company_base_info["REGCAP"]  = company_base_info["REGCAP"].fillna(0)
company_base_info["REGCAPCUR"]  = company_base_info["REGCAPCUR"].fillna(156)
company_base_info["RECCAP"]  = company_base_info["RECCAP"].fillna(0)
company_base_info["certificate_num"]  = company_base_info["certificate_num"].fillna(0)
company_base_info["brand_num"]  = company_base_info["brand_num"].fillna(0)

company_base_info["cap_ratio"]  = 0
company_base_info.loc[company_base_info["REGCAP"] != 0, "cap_ratio"] = company_base_info["RECCAP"]/company_base_info["REGCAP"]
company_base_info.loc[company_base_info["cap_ratio"] >= 1, "cap_ratio"] = 1
```

```python
caec = pd.get_dummies(company_base_info["REGCAPCUR"], prefix="REGCAPCUR")
company_base_info = pd.concat([company_base_info, caec],axis=1)
company_base_info
```

```python
dic_cur = {156.0: 1, 840.0: 6.48, 344.0: 0.83, 954.0: 7.64,       
           36.0: 0.087, 392.0: 0.059, 4.0: 9.15, 978.0: 6.48,
           170.0: 0.0017, 0: 1}
company_base_info["REGCAP_Normal"]  = company_base_info.apply(lambda row: row['REGCAP'] * dic_cur[row['REGCAPCUR']], axis = 1)
company_base_info["RECCAP_Normal"]  = company_base_info.apply(lambda row: row['RECCAP'] * dic_cur[row['REGCAPCUR']], axis = 1)
```

```python
company_base_info = company_base_info.drop(["ENTTYPE","INDUSTRYPHY","PARNUM","LIMPARNUM","PARFORM","EXENUM","REGCAP","REGCAPCUR","RECCAP"],axis=1)
```

```python
company_base_info.to_csv(outpath+"company_base_info_processed.csv",index=False)
```
