```python
import pandas as pd
import numpy as np
```

```python
path = "../Data/Online_Data/trainning_data/corporate_attributes/"
outpath = "../Processed_Data/"
```

```python
company_modify = pd.read_csv(path+"company_modify.csv")
```

```python
company_modify
```

```python
company_modify = company_modify.dropna(subset=["ALTDATE"])
company_modify["ALTDATE"] = company_modify["ALTDATE"].astype(str)
company_modify["year"] = company_modify["ALTDATE"].str[:4]
company_modify["year"] = company_modify["year"].astype(int)
```

```python
company_modify = company_modify.fillna({'ALTTIME': 0})
company_modify
```

```python
df_login = company_modify.drop(["ALTTIME"],axis=1).copy()
df_login = df_login.groupby(['entid', 'year']).count()
df_login = df_login.groupby(['entid']).mean()
df_login
```

```python
df_alttime = company_modify.groupby(['entid', 'year']).mean().copy()
df_alttime = company_modify.groupby(['entid']).mean()
df_alttime = df_alttime.drop(["year"],axis=1)
df_alttime
```

```python
df_alttime.iloc[:,0]
```

```python
df = pd.DataFrame()
df["nubmer_of_login"] = df_alttime.iloc[:,0]
df["number_of_changes"] = df_login.iloc[:,0]
df
```

```python
df.to_csv(outpath+"company_modify_processed.csv")
```
